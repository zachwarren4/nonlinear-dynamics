```python
import numpy as np
```

```python
def logistic_map(x, r, n):
    x_vals = {0: x}

    for n in np.arange(n):
        x_vals[n+1] = r*x_vals[n]*(1-x_vals[n])
    
    return x_vals
```

```python
logistic_map(.5, 2.5, 3)
```

```python
logistic_map(.2, 2.6, 10)
```

```python
import matplotlib.pyplot as plt
```

```python
prob_3 = logistic_map(.2, 2, 50)
```

```python
plt.plot([int(key) for key in prob_3.keys()], list(prob_3.values()))
```

```python
prob_4 = logistic_map(.2, 2.7, 50)
```

```python
plt.plot([int(key) for key in prob_4.keys()], list(prob_4.values()))
```

```python
def many_maps(f, s_range, r, n):
    fig, ax = plt.subplots(len(s_range), 1, figsize = (7, 20))
    
    results = []
    
    for i, x in enumerate(s_range):
        res = f(x, r, n)
        results.append(res)
        ax[i].scatter([int(key) for key in res.keys()], list(res.values()))
    
    return ax
```

```python
many_maps(logistic_map, [0, .1, .2, .3, .4, .5, .6, .7, 1], 2.8, 50)
```

```python
def many_maps_param(f, s, r_range, n):
    fig, ax = plt.subplots(len(r_range), 1, figsize = (7, 40))
    
    results = []
    
    for i, x in enumerate(r_range):
        res = f(s, x, n)
        results.append(res)
        ax[i].scatter([int(key) for key in res.keys()], list(res.values()))
    
    return res, fig, ax
```

```python
many_maps_param(logistic_map, .2, [2.99, 3, 3.01], 5000)
```

```python
many_maps(logistic_map, [.1*x for x in range(1, 9)], 3.2, 5000)
```

```python
many_maps_param(logistic_map, .2, np.arange(3.5, 3.551, .005), 500)
```
