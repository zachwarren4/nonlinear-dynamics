### Logistic Map
The logistic map is defined as $x_{n+1} = Rx_n(1-x_n)$.

```python
# I use a dictionary just for easy access to values asked for in the course, an array would work fine

def logistic_map(x, r, n):
    x_vals = {0: x}

    for n in np.arange(n):
        x_vals[n+1] = r*x_vals[n]*(1-x_vals[n])
    
    return x_vals
```

### Return Map
A return map is the result of plotting $x_{n+1}$ vs  $x_n$. Here we set up a function to plot that for the logistic map.

```python
def plot_return_map(r):
    
    xn = np.arange(0, 1, .01)
    xn1 = [r*x*(1-x) for x in xn]
    
    fig, ax = plt.subplots(1, 1, figsize = (10, 10))
   
    ax.scatter(xn, xn1, color='red')
    ax.plot(xn[:75], xn[:75])
    return ax
```

```python
plot_return_map(3)
```
